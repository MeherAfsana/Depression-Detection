```python
import shap
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences
from gensim.utils import simple_preprocess
from gensim.models import Word2Vec
import os

# Load the trained GRU model
model_path = r"C:\Users\PC\Downloads\GRU_model_word2vec.h5"  # Adjust path if needed
if not os.path.exists(model_path):
    raise FileNotFoundError(f"Model file not found at {model_path}")
model = load_model(model_path)

# Load dataset
data_path = r"C:\Users\PC\Downloads\your_preprocessed_data.xlsx"  # Adjust path if needed
data = pd.read_excel(data_path)

# Define text column
text_column = "cleaned_text"
data[text_column] = data[text_column].fillna("missing")

# Train a Word2Vec model
print("Training Word2Vec model...")
sentences = data[text_column].apply(simple_preprocess).tolist()
word2vec_model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4, seed=42)

# Reconstruct the word_index dictionary
word_index = {word: i + 1 for i, word in enumerate(word2vec_model.wv.index_to_key)}

# Define max_len
max_len = 100

# Function to preprocess text into Word2Vec sequences
def preprocess_text(text, word_index, max_len):
    sequence = [word_index.get(word, 0) for word in simple_preprocess(text)]
    return pad_sequences([sequence], maxlen=max_len, padding='post', truncating='post')[0]

# Wrapper class for SHAP
class GRUWrapper:
    def __init__(self, model, word_index, max_len):  # Fixed constructor
        self.model = model
        self.word_index = word_index
        self.max_len = max_len
    
    def predict(self, texts):
        sequences = np.array([preprocess_text(text, self.word_index, self.max_len) for text in texts])
        return self.model.predict(sequences)

# Wrap the GRU model
gru_wrapper = GRUWrapper(model, word_index, max_len)

# Select a subset of data for SHAP explanations
background_samples = np.array([preprocess_text(text, word_index, max_len) for text in data[text_column].iloc[:50]])

# SHAP explainer
shap_explainer = shap.Explainer(gru_wrapper.predict, background_samples)

# Generate SHAP values for multiple samples
sample_texts = data[text_column].iloc[:30].tolist()  # Choose 30 samples for better visualization
sample_texts_processed = np.array([preprocess_text(text, word_index, max_len) for text in sample_texts])

shap_values = shap_explainer(sample_texts_processed)

# Convert SHAP values into feature names
feature_names = ["Token_" + str(i) for i in range(max_len)]

# Create SHAP summary plot (beeswarm style)
plt.figure(figsize=(10, 8))
shap.summary_plot(shap_values.values, features=sample_texts_processed, feature_names=feature_names, plot_type="dot", color_bar=True)
plt.title("SHAP Summary Plot")
plt.xlabel("SHAP Value (Impact on Model Output)")
plt.ylabel("Features")
plt.show()
```
